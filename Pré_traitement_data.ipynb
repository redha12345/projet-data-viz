On importe le cours du pétrole et du gaz sur la même période que notre CSV précèdent

```python
import yahoo_fin.stock_info as fi
courtP=fi.get_data("BZ=F", start_date="06-01-2022", end_date="11-11-2022")
courtG=fi.get_data("NG=F",start_date="06-01-2022", end_date="11-11-2022")
```

```python
fi.get_live_price("BZ=F")
```

On retire les colonnes inutiles

```python
courtG.drop(columns=["high","low","low","adjclose","volume","ticker"],inplace=True)
courtP.drop(columns=["high","low","low","adjclose","volume","ticker"],inplace=True)
```

On regroupe les valeurs d'ouverture et de fermeture pour créer la valeur moyenne de nos hydrocarbures sur la journée

```python
courtP=courtP.assign(court=((courtP["open"]+courtP["close"])/2))
courtG=courtG.assign(court=((courtG["open"]+courtG["close"])/2))
```

```python
courtG.drop(columns=["open","close"],inplace=True)
courtP.drop(columns=["open","close"],inplace=True)
```

On crée deux nouveaux CSV avec nos valeurs obtenues

```python
csv_court_gaz_histo=pd.DataFrame.to_csv(courtG,sep= ";")
csv_court_petrol_histo=pd.DataFrame.to_csv(courtP,sep= ";")
courtP.to_csv('csv_court_gaz_histo.csv', sep = ';')
courtG.to_csv('csv_court_petrol_histo.csv', sep = ';')
```
